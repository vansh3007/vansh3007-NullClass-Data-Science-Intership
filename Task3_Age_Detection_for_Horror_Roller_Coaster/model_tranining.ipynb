```python
import os
import gc
import numpy as np
import pandas as pd
from PIL import Image
from tqdm import tqdm
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Input, Conv2D, MaxPooling2D, Flatten, Dense, Dropout
from tensorflow.keras.callbacks import ModelCheckpoint, LearningRateScheduler
from tensorflow.keras.optimizers import Adam


IMAGE_DIR = "../data/UTKFace"
image_paths = []
age_labels = []
gender_labels = []

for filename in tqdm(os.listdir(IMAGE_DIR)):
    try:
        age, gender = filename.split("_")[0:2]
        age = int(age)
        gender = int(gender)
        path = os.path.join(IMAGE_DIR, filename)

        image_paths.append(path)
        age_labels.append(age)
        gender_labels.append(gender)
    except:
        continue


def extract_features(image_paths, size=(96, 96)):  
    features = []
    for img_path in tqdm(image_paths):
        img = Image.open(img_path).convert("L") 
        img = img.resize(size)
        features.append(np.array(img))
    features = np.array(features, dtype=np.float32).reshape(-1, size[0], size[1], 1)
    return features / 255.0

X = extract_features(image_paths)
y_age = np.array(age_labels)
y_gender = np.array(gender_labels)


y_combined = np.column_stack((y_gender, y_age))
X_train, X_temp, y_train_combined, y_temp_combined = train_test_split(X, y_combined, test_size=0.2, random_state=42)


gc.collect()

X_val, X_test, y_val_combined, y_test_combined = train_test_split(X_temp, y_temp_combined, test_size=0.5, random_state=42)

y_train_gender, y_train_age = y_train_combined[:, 0], y_train_combined[:, 1]
y_val_gender, y_val_age = y_val_combined[:, 0], y_val_combined[:, 1]
y_test_gender, y_test_age = y_test_combined[:, 0], y_test_combined[:, 1]


input_shape = (96, 96, 1)  
inputs = Input(shape=input_shape)

x = Conv2D(32, (3, 3), activation='relu')(inputs)
x = MaxPooling2D((2, 2))(x)
x = Conv2D(64, (3, 3), activation='relu')(x)
x = MaxPooling2D((2, 2))(x)
x = Conv2D(128, (3, 3), activation='relu')(x)
x = MaxPooling2D((2, 2))(x)
x = Conv2D(256, (3, 3), activation='relu')(x)
x = MaxPooling2D((2, 2))(x)
x = Flatten()(x)

dense_gender = Dense(256, activation='relu')(x)
dropout_gender = Dropout(0.3)(dense_gender)
output_gender = Dense(1, activation='sigmoid', name='gender_out')(dropout_gender)

dense_age = Dense(256, activation='relu')(x)
dropout_age = Dropout(0.3)(dense_age)
output_age = Dense(1, activation='relu', name='age_out')(dropout_age)

model = Model(inputs=inputs, outputs=[output_gender, output_age])
model.compile(optimizer=Adam(1e-3),
              loss=['binary_crossentropy', 'mae'],
              metrics=['accuracy', 'mae'])

model.summary()


model_path = "best_model.keras"
checkpoint = ModelCheckpoint(model_path, monitor='val_gender_out_accuracy', save_best_only=True, verbose=1, mode='max')
annealer = LearningRateScheduler(lambda epoch: 1e-3 * (0.9 ** epoch))

history = model.fit(
    X_train, [y_train_gender, y_train_age],
    validation_data=(X_val, [y_val_gender, y_val_age]),
    epochs=20,
    batch_size=128,
    callbacks=[checkpoint, annealer]
)


plt.plot(history.history['gender_out_accuracy'], label='Gender Accuracy (Train)')
plt.plot(history.history['val_gender_out_accuracy'], label='Gender Accuracy (Val)')
plt.legend()
plt.title("Gender Accuracy")
plt.show()

plt.plot(history.history['age_out_mae'], label='Age MAE (Train)')
plt.plot(history.history['val_age_out_mae'], label='Age MAE (Val)')
plt.legend()
plt.title("Age MAE")
plt.show()

model.save("age_gender_model.h5")
```
